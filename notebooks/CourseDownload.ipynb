```python
input_file = open('../data/universities_list.pkl', 'rb')
universities = pkl.load(input_file)
```

```python
university_list = []
for university in universities:
    university = university.split('-')
    if university[-1].isnumeric():
        temp_dict = {}
        temp_dict['id'] = university[-1]
        temp_dict['university_name'] = ' '.join(university[:-1])
        university_list.append(temp_dict)
```

```python
course_list = []
count = 0
for university in university_list:
    count += 1
    course_link = "https://api.yocket.com/universities/courses/"+university['id']+"/0/e242f222-9e4d-4816-9d43-e75e71fb4363?is_review=0&page=0&items=1000"
    html = urllib.request.urlopen(course_link)
    text = html.read()
    y = json.loads(text)
    for course in y['data']['courses']['results']:
        course['university_id']= university['id']
        course_list.append(course)
    print('{}. University ID : {} Name : {} Courses Added : {}'.format(count, university['id'], university['university_name'], len(y['data']['courses']['results'])))
```
